```python
import pandas as pd
import sqlite3

# Load CSV
df = pd.read_csv(r"C:\Users\nanin\OneDrive\Desktop\intrn_s\anime\popular_anime.csv")

# Push to SQLite
conn = sqlite3.connect("anime.db")
df.to_sql("anime", conn, if_exists="replace", index=False)
print("✅ Loaded into SQLite!")

# Query 1 — Total records
print(pd.read_sql("SELECT COUNT(*) AS total FROM anime", conn))
```

```python
# Query 2 — Top 10 highest rated anime
print(pd.read_sql("""
    SELECT name, score, rank, genres
    FROM anime
    WHERE score IS NOT NULL
    ORDER BY score DESC
    LIMIT 10
""", conn))
```

```python
# Query 3 — Count by type (TV, Movie, OVA etc.)
print(pd.read_sql("""
    SELECT type, COUNT(*) AS count
    FROM anime
    GROUP BY type
    ORDER BY count DESC
""", conn))
```
